In this project we will work with [a dataset on the job outcomes of students who graduated from college from 2010 to 2012](https://github.com/fivethirtyeight/data/tree/master/college-majors). We will create visualization to answer questions regarding college graduates and their majors

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
```

```python
recent_grads.head()
```

```python
recent_grads.tail()
```

```python
recent_grads.describe()
```

```python
raw_data_count = recent_grads.shape
raw_data_count
```

```python
recent_grads=recent_grads.dropna()
```

```python
cleaned_data_count = recent_grads.shape
cleaned_data_count
```

```python
recent_grads.plot(x='Sample_size',
                  y='Median', kind='scatter', 
                  title='Sample size of full time employees vs Median Salary',
                  figsize=(5,10))
```

```python
recent_grads.plot(x='Sample_size',
                  y='Unemployment_rate', kind='scatter', 
                  title='Sample size of full-time vs Unemployment rate',
                  figsize=(5,10))
```

```python
recent_grads.plot(x='Full_time',
                  y='Median', kind='scatter', 
                  title='Number of full-time workers vs Median salary',
                  figsize=(5,10))
```

```python
recent_grads.plot(x='ShareWomen',
                  y='Unemployment_rate', kind='scatter', 
                  title='Percentage of women vs Unemployment rate',
                  figsize=(5,10))
```

```python
recent_grads.plot(x='Men',
                  y='Median', kind='scatter', 
                  title='Male graduates vs Median Salary',
                  figsize=(5,10))
```

```python
recent_grads.plot(x='Women',
                  y='Median', kind='scatter', 
                  title='Female graduates vs Median Salary',
                  figsize=(5,10))
```

From these initial visualizations we can see some correlations between the columns of the dataset
